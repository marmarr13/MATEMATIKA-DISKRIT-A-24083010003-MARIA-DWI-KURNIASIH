```python
def cek_pelesetan(n):
    # Pernyataan 1: n^2 - n habis dibagi (n-1)
    if n > 1:  # biar tidak bagi 0
        pernyataan1 = (n**2 - n) % (n - 1) == 0
    else:
        pernyataan1 = False

    # Pernyataan 2: jumlah deret harmonik < 2
    jumlah = sum(1/i for i in range(1, n+1))
    pernyataan2 = jumlah < 2

    return pernyataan1, pernyataan2, jumlah


# Cetak header tabel
print(f"{'n':<3} | {'Pernyataan1':<12} | {'Pernyataan2':<12} | {'Jumlah Deret':<12}")
print("-"*50)

# Isi tabel untuk beberapa n
for n in range(1, 15):
    p1, p2, jumlah = cek_pelesetan(n)
    print(f"{n:<3} | {str(p1):<12} | {str(p2):<12} | {jumlah:<12.4f}")

import matplotlib.pyplot as plt

# Hitung jumlah deret harmonik
n_values = list(range(1, 51))
harmonic_sums = [sum(1/i for i in range(1, n+1)) for n in n_values]

# Plot
plt.figure(figsize=(8,5))
plt.plot(n_values, harmonic_sums, marker='o', label="Jumlah Deret (Harmonik)")
plt.axhline(y=2, color='r', linestyle='--', label="Batas = 2")
plt.title("Visualisasi Jumlah Deret Harmonik")
plt.xlabel("n (banyaknya suku)")
plt.ylabel("Jumlah Deret")
plt.legend()
plt.grid(True)
plt.show()
```

Berdasarkan hasil yang telah diperoleh, n yang sesuai untuk pernyataan pertama yaitu n^2-n = n(n-1) adalah n>1 dengan n tidak sama dengan 1.
Kemudian untuk pernyataan kedua mengenai deret, berdasarkan hasil visualisasi yang telah diperoleh, n yang masih memenuhi adalah n<4, n>4 sudah tidak sesuai karena semakin bertambahnya n, deret hasil yang diperoleh juga semakin besar dan lebih dari 2. Untuk yang masih memenuhi kurang dari 2 adalah n<4.
