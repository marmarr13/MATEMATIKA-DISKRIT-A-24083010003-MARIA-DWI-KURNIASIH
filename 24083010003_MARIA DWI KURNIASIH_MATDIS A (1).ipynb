```python
#import library yang akan dibutuhkan
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#data yang diambil pada csv merupakan data random (penilaian acak)
data = pd.read_csv("Downloads/class.csv") #inisialisasi data
data
```

```python
num_data = data.drop(columns=["NPM", "NAMA"])  # buang kolom yang tidak penting 
num_data
```

```python
cov_mat = np.cov(num_data.T)

# Matriks korelasi
corr_mat = np.corrcoef(num_data.T)

# Ambil nama kolom
cols = num_data.columns

# Matriks kovarian dengan label
cov_data = pd.DataFrame(cov_mat, index=cols, columns=cols)

# Matriks korelasi dengan label
corr_data = pd.DataFrame(corr_mat, index=cols, columns=cols)

print("Matriks Kovarian:")
print(cov_data, "\n")

print("Matriks Korelasi:")
print(corr_data)
```

# MATRIKS KOVARIAN

Pada hasil matriks kovarian diatas menunjukkan bahwa variabel dengan nilai paling bervarian atau penyebaran data yang luas adalah variabel "ADIL DAN AMANAH" dengan nilai sebesar 0,099236 dan variabel dengan nilai varian yang rendah atau penyebaran data yang sedikit adalah variabel "NYAMBUNG KALAU NGOBROL". 

Variabel yang memiliki arah berlawanan yaitu "ADIL DAN AMANAH" dengan "SUKA MEMBANTU", "NYAMBUNG KALAU NGOBROL" dengan "SUKA MEMBANTU", dan "ASYIK DAN MENARIK" dengan "SUKA MEMBANTU". Hal ini berarti, ketika satu nilai variabel tinggi maka nilai variabel lainnya akan rendah ataupun sebaliknya.

Pada variabel yang memiliki nilai positif satu sama lain maka kovarian antar variabel memilliki arah yang searah.

# MATRIKS KORELASI
Pada hasil matriks korelasi diatas menunjukkan bahwa seluruh variabel memiliki korelasi yang lemah, maka seluruh variabel memiliki korelasi yang sangat kecil/lemah karena seluruh variabel memiliki nilai r antara 0-0.3 kecuali korelasi dengan variabelnya sendiri.

Kemudian, nilai yang bertanda negatif menunjukkan bahwa variabel tersebut berlawanan arah, contohnya variabel "ASYIK DAN MENARIK" dengan "SUKA MEMBANTU". Variabel yang tidak bernilai negatif berarti kedua variabel bergerak searah, saat satu variabel naik, variabel lain akan naik contoh variabel yang searah adalah variabel "ADIL DAN AMANAH" dengan "NYAMBUNG KALAU NGOBROL".


```python
cols = num_data.columns  # nama variabel numerik

# Visualisasi matriks Kovarian
plt.figure(figsize=(6, 4))
sns.heatmap(cov_mat, xticklabels=cols, yticklabels=cols,
            annot=True, fmt=".3f", cmap="coolwarm")
plt.title("Heatmap Kovarian")
plt.show()

# Visualisasi matriks Korelasi
plt.figure(figsize=(6, 4))
sns.heatmap(corr_mat, xticklabels=cols, yticklabels=cols,
            annot=True, fmt=".3f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Heatmap Korelasi")
plt.show()
```

Berdasarkan hasil visualisasi matriks kovarian dan matriks korelasi menggunakan heatmap, semakin berwarna merah maka hubungan antar variabel semakin kuat dan semakin berwarna biru maka hubungan antar variabel sangat lemah. Angka negatif dan positif hanya untuk menunjukkan arah.

```python
# 2. Ubah ke numpy array
X_matrix = num_data.values

# 3. Hitung cosine similarity
norms = np.linalg.norm(X_matrix, axis=1, keepdims=True)
cosine_sim = (X_matrix @ X_matrix.T) / (norms @ norms.T)

# 4. Ambil kolom NAMA hanya untuk label (index/columns)
students = data["NAMA"].values

# 5. Simpan ke DataFrame biar rapi
cosine_df = pd.DataFrame(cosine_sim, index=students, columns=students)
cosine_df.to_csv("cosine_similarity.csv", index=True)
```

```python
# Visualisasi heatmap cosine similarity
plt.figure(figsize=(10,8))
sns.heatmap(cosine_df, annot=False, cmap="coolwarm")

plt.title("Heatmap Cosine Similarity antar Mahasiswa")
plt.show()
```

Cosine Similarity merupakan sebuah metode untuk mengukur kemiripan nilai antar variabel atau antar vektor dimana pada data diatas tiap mahasiswa sudah diberikan nilai masing-masing pada variabel. 

Berdasarkan hasil visualisasi cosine similarity diatas, semakin berwarna merah atau mendekati nilai 1, maka mahasiswa satu dengan mahasiswa yang lain memiliki hasil penilaian yang mirip dalam seluruh variabel. Namun, semakin berwarna biru atau kurang dari 0.2 maka mahasiswa satu dengan mahasiswa yang lain tidak memiliki kemiripan hasil penilaian dalam seluruh variabel. Untuk mahasiswa dengan nama yang sama (dibandingkan dengan dirinya sendiri) sudah dipastikan akan menepati nilai cosine sebesar 1.

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# --- PCA 2 dimensi untuk visualisasi ---
pca = PCA(n_components=2)
X_pca = pca.fit_transform(num_data)

# --- KMeans clustering ---
kmeans = KMeans(n_clusters=3, random_state=42)  # pilih jumlah cluster
clusters = kmeans.fit_predict(num_data)

# Tambahkan hasil cluster ke dataframe
data["Cluster"] = clusters

# --- Visualisasi PCA + Cluster ---
plt.figure(figsize=(10,8))
scatter = plt.scatter(X_pca[:,0], X_pca[:,1], c=clusters, cmap="Set2", edgecolor="k")

# Kasih label nama mahasiswa
for i, name in enumerate(data["NAMA"]):
    plt.text(X_pca[i,0]+0.02, X_pca[i,1]+0.02, name, fontsize=8)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA + KMeans Clustering Mahasiswa")
plt.colorbar(scatter, label="Cluster")
plt.show()
```

metode PCA (Principal Component Analysis) merupakan metode untuk merangkum banyak variabel misalnya PC1 dan PC2 supaya data bisa divisualisasikan dengan sederhana. Output diatas merupakan hasil dari PCA serta clustering menggunakan K-Means yang akan mengelompokkan data-data mahasiswa yang mendapatkan penilaian mirip atau kemiripan karakter. 

Jika hanya melihat pada visualisasi cosine similarity masih tidak efisien jika harus melihat satu per-satu mahasiswa yang memiliki kemiripan karakter, maka dilakukan clustering supaya terlihat jelas beberapa mahasiswa yang memiliki kemiripan karakter. 

Berdasarkan visualisasi diatas menunjukkan bahwa setiap nama yang memiliki kesamaan warna, maka mereka memiliki kemiripan karakter.

```python
for c in sorted(data["Cluster"].unique()):
    print(f"\nCluster {c}:")
    print(data.loc[data["Cluster"] == c, "NAMA"].tolist())
```

Output diatas merupakan hasil dari clustering dalam 3 cluster. Cluster 0 menunjukkan 18 mahasiswa, cluster 1 menunjukkan 14 mahasiswa, dan cluster 2 menunjukkan 8 mahasiswa. Setiap cluster memiliki nama-nama mahasiswa yang memiliki rata-rata penilaian yang mirip.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Kolom numerik yang dipakai untuk klasifikasi
num_cols = ["ADIL DAN AMANAH", "SUKA MEMBANTU", 
            "NYAMBUNG KALAU NGOROL", "ASYIK DAN MENARIK"]

# Ambil fitur X
X = data[num_cols]

# --- Bikin label berdasarkan rata-rata ---
data["avg_score"] = X.mean(axis=1)
data["Label"] = pd.cut(data["avg_score"],
                     bins=[-0.01, 0.4, 0.7, 1.0],
                     labels=["Rendah", "Sedang", "Tinggi"])

y = data["Label"]

# --- Split data train/test ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# --- Standardisasi data ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

# --- Definisikan model KNN ---
clf = KNeighborsClassifier(n_neighbors=5)

# Training
clf.fit(X_train_scaled, y_train)

# --- Evaluasi ---
print("Akurasi di data train:", clf.score(X_train_scaled, y_train))
print("Akurasi di data test :", clf.score(X_test_scaled, y_test))

# --- Prediksi untuk seluruh data (pakai scaler juga) ---
X_scaled = scaler.transform(X)
all_pred = clf.predict(X_scaled)

hasil_all = pd.DataFrame({
    "Nama": data["NAMA"],
    "Label Asli": data["Label"],
    "Prediksi": all_pred
})

# Tambahkan kolom cek benar/salah
hasil_all["Benar/Salah"] = hasil_all["Label Asli"] == hasil_all["Prediksi"]
hasil_all["Benar/Salah"] = hasil_all["Benar/Salah"].map({True: "✓", False: "✗"})

# Tampilkan semua hasil
pd.set_option("display.max_rows", None)
print("\nHasil Prediksi:")
print(hasil_all)
```

Output diatas merupakan hasil dari prediksi dengan mengklasifikasikan nilai rata-rata mahasiswa menjadi rendah, sedang, dan tinggi. Nilai rendah ada pada angka -0,01<nilai<=0,4. Nilai sedang ada pada angka 0,4<nilai<=0,7 serta nilai yang tinggi pada angka 0,7<sampai<=1. Label asli merupakan nilai rata-rata sebenarnya dengan rentang yang sudah ditentukan, kemudian seluruh data akan diprediksi apakah sama dengan label asli (nilai asli). Hasil menunjukkan bahwa terdapat 10 mahasiswa yang memiliki hasil prediksi tidak sesuai dengan label asli. Akurasi data trainn dengan data test juga memiliki akurasi model yang cukup baik atau stabil yaitu 0.75

Klasifikasi prediksi menggunakan  KNN (K-Nearest Neighbors) dilakukan untuk mengetahui seberapa baik/seberapa akuratnya model mengenali pola nilai karakteristik mahasiswa. Dalam menggunakan metode KNN, data akan distandarisasi dulu menggunakan StandardScaler untuk mencegah sensitifnya perbedaan skala antar variabel.
