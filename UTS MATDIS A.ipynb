```python
from docx import Document
import re
import pandas as pd
import matplotlib.pyplot as plt

# === FUNGSI UNTUK MENGHITUNG KONSONAN & VOKAL ===
def count_vocals_consonants(word):
    word = word.lower()
    vocals = sum(1 for c in word if c in "aeiou")
    consonants = sum(1 for c in word if c.isalpha() and c not in "aeiou")
    return consonants, vocals


# === 1️⃣ BACA FILE WORD (.DOCX) ===
doc_path = "DeTi A Decentralized Ticketing Management Platform Clean.docx"
doc = Document(doc_path)

# === 2️⃣ AMBIL TEKS, KECUALIKAN YANG BERGAYA HEADING ===
paragraphs = []
for para in doc.paragraphs:
    text = para.text.strip()
    if not text:
        continue
    # Jika gaya paragraf termasuk heading, lewati
    if para.style.name.lower().startswith("heading"):
        continue
    paragraphs.append(text)

# Gabungkan semua paragraf isi (tanpa heading)
full_text = " ".join(paragraphs)

# === 3️⃣ PEMBERSIHAN TEKS ===
text_clean = re.sub(r"[^a-zA-Z\s-]", " ", full_text)
text_clean = re.sub(r"\s+", " ", text_clean).strip().lower()

# === 4️⃣ EKSTRAK SEMUA KATA ===
words = re.findall(r"\b[a-z]+\b", text_clean)

# === 5️⃣ TABULASI PER KATA ===
unique_words = sorted(set(words))
data = [
    {"word": w, "consonants": count_vocals_consonants(w)[0], "vocals": count_vocals_consonants(w)[1]}
    for w in unique_words
]
df_words = pd.DataFrame(data)

# === TOTAL SELURUH VOKAL DAN KONSONAN ===
total_consonants = df_words["consonants"].sum()
total_vocals = df_words["vocals"].sum()
total_letters = total_consonants + total_vocals

# === JUMLAH KATA ===
total_unique_words = len(df_words)       # jumlah kata unik
total_all_words = len(words)             # jumlah semua kata termasuk duplikat dari teks asli

print("\n=== REKAPITULASI HURUF & KATA ===")
print(f"Total konsonan                 : {total_consonants}")
print(f"Total vokal                    : {total_vocals}")
print(f"Total keseluruhan huruf        : {total_letters}")
print(f"Jumlah kata unik               : {total_unique_words}")
print(f"Jumlah seluruh kata (termasuk duplikat): {total_all_words}")

print("\n=== TABULASI PER KATA (INDIVIDUAL) ===")
print(df_words.head(15))
```

```python
from docx import Document
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# === 2️⃣ GABUNGKAN BULLET DAN KALIMAT SETELAHNYA KE PARAGRAF SEBELUMNYA ===
merged_paragraphs = []
temp = ""
after_bullet = False  # flag penanda kalau kita baru lewat bullet

for para in doc.paragraphs:
    text = para.text.strip()
    if not text:
        continue
    if para.style.name.lower().startswith("heading"):
        continue

    # Kalau bullet point
    if text.startswith(("•", "-", "*")):
        temp += " " + text
        after_bullet = True
    # Kalau paragraf setelah bullet (masih penjelasan yang sama)
    elif after_bullet and not re.match(r"^(abstract|introduction|summary|discussion)", text.lower()):
        temp += " " + text
        after_bullet = False  # reset flag
    else:
        # simpan paragraf sebelumnya
        if temp:
            merged_paragraphs.append(temp.strip())
        temp = text
        after_bullet = False

# tambahkan paragraf terakhir
if temp:
    merged_paragraphs.append(temp.strip())

paragraphs = merged_paragraphs

print(f"Total paragraf (setelah digabung bullet dan kelanjutannya): {len(paragraphs)}")

# === 3️⃣ HITUNG JUMLAH KALIMAT PER PARAGRAF ===
paragraph_sentence_counts = []
for para in paragraphs:
    sentences_in_para = re.split(r'[.!?]+', para)
    sentences_in_para = [s.strip() for s in sentences_in_para if s.strip()]
    paragraph_sentence_counts.append(len(sentences_in_para))

# === 4️⃣ BUAT DATAFRAME UNTUK ANALISIS ===
df_paragraphs = pd.DataFrame({
    "paragraph": paragraphs,
    "sentence_count": paragraph_sentence_counts
})

print("\n=== CONTOH 10 PARAGRAF PERTAMA DENGAN JUMLAH KALIMAT ===")
print(df_paragraphs.head(10))

# === 5️⃣ STATISTIK DESKRIPTIF ===
print("\nStatistik jumlah kalimat per paragraf:")
print(df_paragraphs["sentence_count"].describe())

# === 6️⃣ VISUALISASI HISTOGRAM ===
plt.figure(figsize=(10,6))
sns.histplot(df_paragraphs["sentence_count"], bins=15, kde=True, color="lightcoral", edgecolor="black")
plt.title("Distribusi Jumlah Kalimat per Paragraf (Bullet dan Kelanjutannya Digabung)")
plt.xlabel("Jumlah Kalimat dalam Paragraf")
plt.ylabel("Frekuensi")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

Berdasarkan output yang telah didapat, rata-rata per 1 paragaraf memiliki kalimat sejumlah kisaran 10 sampai 12 kalimat sehingga menunjukkan bahwa setiap paragraf memiliki isi yang cukup informatif. Terdapat juga paragraf yang mmemiliki jumlah kalimat hingga 22 kalimat yang menunjukkan penjelasan lebih mendalam pada paragraf tersebut.
